```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("world_population.csv")
df.head()
```

```python
# Set the chart style
sns.set_style("whitegrid")

# Tell Seaborn to create the bar chart
sns.countplot(x="Continent", data=df, order=df['Continent'].value_counts().index)

# Add titles
plt.title("Number of Countries in Each Continent")
plt.xlabel("Continent")
plt.ylabel("Number of Countries")

# Rotate the labels on the x-axis to make them readable
plt.xticks(rotation=45)

# Make sure everything fits nicely
plt.tight_layout()

# Show the chart
plt.show()
```
